```python
import time

# Fungsi untuk Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    start_time = time.time()

    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    end_time = time.time()
    execution_time = end_time - start_time

    return execution_time

# Fungsi untuk Insertion Sort
def insertion_sort(arr):
    n = len(arr)
    start_time = time.time()

    for i in range(1, n):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    end_time = time.time()
    execution_time = end_time - start_time

    return execution_time

# Fungsi untuk menampilkan 5 iterasi pertama dan 5 iterasi terakhir
def display_iterations(arr):
    print("5 Iterasi Pertama:")
    for i in range(5):
        print(arr[i], end=" ")
    print("\n")

    print("5 Iterasi Terakhir:")
    n = len(arr)
    for i in range(n - 5, n):
        print(arr[i], end=" ")
    print("\n")

# Fungsi untuk menampilkan waktu komputasi pengurutan
def display_execution_time(execution_time):
    print("Waktu Komputasi: {:.6f} detik".format(execution_time))

# Fungsi untuk menampilkan array sebelum dan sesudah pengurutan
def display_before_after(arr, sorted_arr):
    print("Sebelum Pengurutan: ")
    print(arr)
    print("\nSesudah Pengurutan: ")
    print(sorted_arr)
    print("\n")

# Fungsi untuk menganalisis algoritma
def analyze_algorithm():
    print("Analisis Algoritma:")
    print("1. Worst Case: O(n^2)")
    print("2. Best Case: O(n)")
    print("3. Average Case: O(n^2)")

# Main Program
def main():
    arr = [12, 99, 62, 15, 20, 95, 39, 48, 3, 24, 8, 43, 74, 19, 97, 33, 49, 68, 55, 33, 90, 90, 7,
           26, 85, 46, 39, 40, 9, 36, 60, 64, 89, 31, 25, 71, 21, 23, 63, 84, 32, 5, 3, 44, 21, 10, 21,
           17, 50, 2, 36, 53, 79, 54, 19, 88, 1, 32, 31, 15, 6, 3, 1, 40, 22, 43, 18, 1, 77, 9, 59,
           40, 7, 41, 81]

    print("=================================")
    print("=        Rivaldi Sabala         =")
    print("=          F55121039            =")
    print("=           Kelas B             =")
    print("=================================")
    print("\n\nPilihan Algoritma:")
    print("1. Bubble Sort")
    print("2. Insertion Sort")
    print("3. Analisis Algoritma")
    choice = input("Masukkan pilihan Anda (1/2/3): ")

    if choice == "1":
        sorted_arr = arr.copy()
        execution_time = bubble_sort(sorted_arr)
        display_iterations(sorted_arr)
        display_execution_time(execution_time)
        display_before_after(arr, sorted_arr)
    elif choice == "2":
        sorted_arr = arr.copy()
        execution_time = insertion_sort(sorted_arr)
        display_iterations(sorted_arr)
        display_execution_time(execution_time)
        display_before_after(arr, sorted_arr)
    elif choice == "3":
        analyze_algorithm()
    else:
        print("Pilihan tidak valid.")

if __name__ == '__main__':
    main()
```
